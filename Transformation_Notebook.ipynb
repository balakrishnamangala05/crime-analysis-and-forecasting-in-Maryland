```python
import pandas as pd

# Load the dataset
data_path = '/content/MD_EduSpending_County.csv'
data = pd.read_csv(data_path)

# Transposing the dataset while keeping 'Year' intact
# First, let's melt the dataframe
df_melted = data.melt(id_vars=['Year', 'Grades'], var_name='County', value_name='Enrollment')

# Now pivot the table to get the desired format
df_pivot = df_melted.pivot_table(index=['County', 'Year'], columns='Grades', values='Enrollment', aggfunc='first').reset_index()

# Reorder the columns as specified
column_order = ['County', 'Year', 'Grades Pre-K', 'Grades K-5', 'Grades 6-8', 'Grades 9-12']
df_final = df_pivot[column_order]

df_final.head()

```
